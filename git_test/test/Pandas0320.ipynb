### <font face="Times New Roman" size=4 > (一)空值處理 </font>

https://medium.com/@zector1030/pandas-fillna-%E7%AF%84%E4%BE%8B-5d33819fb7b8

```python
'''
利用 fillna 函式將 DataFrame 裡頭所有不存在的值設為 0
或是可以設定成自己容易辨別的字串:

df.fillna("Unknown")

'''

df['siteaddress'].fillna(value="Unknown", inplace=True)
df.head()
```

```python
'''
如果是數值型資料可以用平均值取代 nan (或是其他方法)

'''

df['欄位'].fillna(value=df['欄位'].mean(), inplace=True)
df
```

### <font face="Times New Roman" size=4 > (二)取得想要的數據 </font>

*   <font face="Times New Roman" size=4 >loc 含頭也含尾</font>
*   <font face="Times New Roman" size=4 >iloc 含頭不含尾</font>

> column name 取資料的時候用 loc，用 index 取資料的時候用 iloc

https://ticyyang.medium.com/python%E8%B3%87%E6%96%99%E5%88%86%E6%9E%90-pandas%E7%9A%84-loc%E8%88%87iloc%E4%B9%8B%E7%95%B0%E5%90%8C-bf7d62541676

```python
# 例如我要取第 2~5 的 row：
df_loc = df.loc[2:5]
print(df_loc)

# df_iloc = df.iloc[2:6]
# print(df_iloc)
```

```python
# 以column name取資料

# column name沒加上中括號，資料格式為Series
df.loc[:, 'township']    

# column name加上中括號，資料格式為DataFrame
df.loc[:, ['township']]    

#取多個column
df.loc[:, ['areaname', 'township', 'siteid']]
```

```python
# 以index取資料

# index沒加上中括號，資料格式為Series
df.iloc[:, 2]     

# index加上中括號，資料格式為DataFrame
df.iloc[:, [2]]     

# 取多個 column
# 1.連續
df.iloc[:, 2:4] 

# 2.不連續
df.iloc[:, [2, 6, 4]]  
```

### <font face="Times New Roman" size=4 > (四)切割成多個欄位 </font>

```python
# 字串

df = pd.DataFrame({
    "name": ["大雄", "胖虎"], 
    "feature": ["膽小, 翻花繩", "粗魯, 演唱會"]
})
df
```

```python
df[['性格', '特技']] = df.feature.str.split(',', expand=True)
df
```

```python
#list
df = pd.DataFrame({
    "name": ["大雄", "胖虎"], 
    "feature": [["膽小", "翻花繩"], ["粗魯", "演唱會"]]
})

df
```

```python
cols = ['性格', '特技']
df_feature = pd.DataFrame(df.feature.tolist(), columns=cols)

df_name = df.iloc[:,[0]]
# df = pd.concat([df_name, df_feature], join='outer', ignore_index=True)
df = pd.concat([df_name, df_feature], axis=1)

df
```

```python
#也可以用apply來切割

df.feature.apply(pd.Series)
```
